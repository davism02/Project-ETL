```python
import pandas as pd
```

```python
# Import the csv file
csv_1 = 'Resources/2016poverty.csv'
poverty_2016 = pd.read_csv(csv_1)
```

```python
#Check the contents
poverty_2016
```

```python
#Remove Puerto Rico since other data doesn't include it and technically it is not a state.
poverty_2016 = poverty_2016.loc[:51]
```

```python
# Remove the Footnotes column
poverty_2016 = poverty_2016.drop(columns=['Footnotes'])
```

```python
# Put 0s in place of NaNs
poverty_2016 = poverty_2016.fillna(0)
poverty_2016.head(10)
```

```python
# Import the csv file and do the same modifications like we did to the 2016 file
poverty_2016.to_csv('Resources/poverty_2016_clean.csv')
```

```python
csv_2 = 'Resources/2017poverty.csv'
poverty_2017 = pd.read_csv(csv_2)
```

```python
poverty_2017 = poverty_2017.loc[:51]
poverty_2017 = poverty_2017.drop(columns=['Footnotes'])
poverty_2017 = poverty_2017.fillna(0)
poverty_2017.head(10)
```

```python
poverty_2017.to_csv('poverty_2017_clean.csv')
```

```python
#Import population csv file into a dataframe
csv_3 = 'Resources/population.csv'
population = pd.read_csv(csv_3)
population.head()
```

```python
# Filter out the following four columns that we would like to keep and change the names of the columns
population_df = population[['rank', 'State', 'Pop2018', 'density']]
population_df = population_df.rename(columns={'State': 'state', 'Pop2018': 'population', 'density': 'density_per_sq_mi'})
```

```python
population_df.head()
```

```python
#Save to a new clean csv file
population_df.to_csv('population_clean.csv')
```
